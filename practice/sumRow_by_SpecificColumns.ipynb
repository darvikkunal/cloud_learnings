```python
import pandas as pd
import numpy as np
```

```python
#create a pandas dataframe
data = {'col1': ['john', 'sally', 'dorthy', 'cindy', 'mark'],
        'col2': [1,2,5,7,0],
        'col3': [1,np.nan,2,7,10]
       }
df = pd.DataFrame(data=data)
```

```python
#see top 5 rows
df.head()
```

```python
#sum rows for col2 and col3
dflist = ['col2','col3']
df['results'] = df[dflist].sum(axis=1, skipna=False)
df
```
