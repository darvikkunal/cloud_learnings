```python
# Nested loops : tables --> columns (using Titanic df)

import pandas as pd
df = pd .read_csv("/Users/darvikkunalbanda/DATA_ENGINEERING/cloud_learnings/data/titanic.csv")

table_columns = {
    "passengers": ["name", "age", "fare"],
    "id_only": ["survived"]  # Use an existing column
}

for table, cols in table_columns.items():
    print(f"\nTable: {table}")
    for col in cols:
        nulls = df[col].isnull().sum()
        print(f"  {col}: {nulls} nulls")
```

```python
# nested loops with if/else

rules = {
    "string": ["name", "sex"],
    "numeric": ["age", "fare"]
}

for dtype , cols in rules.items():
    #print(dtype , cols)
    print(f"\n Checking {dtype} columns : ")
    for col in cols:
        nulls = df[col].isnull().sum()
        if nulls > 100:
            print(f" {col} : {nulls} nulls --> need attention")
        else:
            print(f" {col} : {nulls} nulls --> OK")
```
