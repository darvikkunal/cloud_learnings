Exercise 2: String Methods on Titanic Names

```python
# Inspect messy names
print("Before : ")
print(df["name"].head())
```

```python
#Clean : strip spaces + lowercase
df['clean_names'] = df["name"].str.strip().str.lower()
print("\n After: ")
print(df['clean_names'].head())
```

```python
# Check a specific row
messy = df["name"][0]
print(messy)
messy_lower = messy.strip().lower()
print(f"\n Messy : '{messy}' --> Clean : '{messy_lower}' ")
messy_proper = messy.title()
print(f"\n Messy : '{messy}' --> Clean : '{messy_proper}' ")
```
